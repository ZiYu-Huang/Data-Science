```python
import pandas as pd
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/ZiYu-Huang/Practices-for-Beginners/main/train.csv'
data_set = pd.read_csv(url)
```

#長條圖 barplot、countplot
通常用來觀察不同類別對於相同屬性值的呈現結果


長條圖的x軸為**類別型**的資料，例:XX的有無、屬於什麼類別

##barplot

```python
sns.barplot(x=[1,2,3,4,5,6,7,8],y=[2,4,6,8,10,12,14,16]) #最普通的用法:在參數x、y傳入序列物件
```

```python
sns.barplot(x='Pclass',y="Survived",data=data_set)
#配合pandas:
#Pclass表示所屬艙等，分為1、2、3，而此時我們把這視為不同的類別
#Survive表示有無生存，0表示無、1表示生存，此時我們把這視為想觀察的相同屬性值
#x、y此兩參數分別填入x、y軸想表達的數值欄位，data傳入所使用的DataFrame
```

```python
sns.barplot(x='Pclass',y="Survived",hue='Sex',data=data_set) #可在參數heu填入欄位來做更進一步的分類
```

##countplot

```python
sns.countplot(x='Pclass',data=data_set) #觀察各Pclass(艙等)的數量
```

```python
sns.countplot(x='Survived',data=data_set) #觀察各Survived(有無生存)的數量
```

```python
sns.countplot(x='Survived',hue='Sex',data=data_set)
```

```python
data_set.info()
```

#直方圖histplot
通常用來觀察**連續性**資料對於相同屬性值的呈現結果


直方圖的x軸為連續性資料，例:年齡、年份等

```python
sns.histplot(x='Age',data=data_set) #可傳x與y參數，如果不傳y參數，那y軸就是數量
```

```python
sns.histplot(x='Age',hue='Survived',data=data_set) #放入參數hue做分類來觀察
```

#散佈圖jointplot
通常用來觀察兩資料之間的關係及分布情形

```python
import requests
import io
url = 'https://github.com/1010code/iris-dnn-tensorflow/raw/master/data/Iris.csv' #用鳶尾花資料集做示範
s=requests.get(url).content
iris_data=pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
iris_data.info() 
#資料集介紹:
#SepalLengthCm(萼片長度)、SepalWidthCm(萼片寬度)、PetalLengthCm(花瓣長度)
#PetalWidthCm(花瓣寬度)、Species(屬於哪一種花)
```

```python
sns.jointplot(x='SepalLengthCm',y='SepalWidthCm',hue="Species",data=iris_data)
```

#盒鬍圖boxplot
可用來觀察資料的分布情形


最大值、最小值、中位數、四分位數

```python
data_set.info()
```

```python
sns.boxplot(x='Survived',y='Age',data=data_set)
'''
長方形最上面那條線為第三四分位數，資料從小到大的第75%資料
長方形中間那條線為第二四分位數，資料從小到大的第50%資料，也就是中位數
長方形最下面那條線為第一四分位數，資料從小到大的第25%資料
最上面以及最下面分別為最大值、最小值，其他點則為離群值
'''
```

#熱力圖heatmap
可用來觀察數值間的係數關係，***通常會配合pandas的corr***方法使用


corr會算出各值之間的係數關係，會自動過濾掉遺失值或不是數字型態的值


會介於-1~1之間，越靠近1就表示愈成正比，越靠近-1就表示愈成反比，越靠近0表示愈沒有關係

```python
data_corr = data_set.corr()
```

```python
data_corr
```

```python
sns.heatmap(data_corr)
```

```python
sns.heatmap(data_corr,annot=True) #可將參數annot設位Ture，可直接觀察數字
```

```python
sns.pairplot(data_set,hue='Survived') #將Survived放入hue做分類來觀察
```
